```python
import numpy as np
```

```python
# Scenario: A TINY reservoir with 3 possible water levels: [Low, Medium, Full]
# Actions: [Release, Hold]
# Transitions are deterministic for now.
```

```python
states = ["low", "medium", "full"]
actions = ["release", "hold"]

#nested dic


# Transition Table: T[state][action] -> next_state
transitions = {
    "low":    {"release": "low",    "hold": "medium"},
    "medium": {"release": "low",    "hold": "full"},
    "full":   {"release": "medium", "hold": "full"},
}

# Reward Table: R[state][action] -> reward
# low head less power and vv, full hold -> spill
rewards = {
    "low":    {"release": 5,   "hold": 0},
    "medium": {"release": 10,  "hold": 0},
    "full":   {"release": 15,  "hold": -20},
}

# we need to find out the policy that maxes reward

np.random.seed(42)
state = "medium" # assume initial state
total_reward = 0

print(f"{'Step':>4} | {'State':>8} | {'Action':>8} | {'Reward':>6} | {'Next State':>10}")
print("-" * 55)

for step in range(10):
    action = np.random.choice(actions)
    reward = rewards[state][action]
    next_state = transitions[state][action]
    total_reward += reward

    print(f"{step:4d} | {state:>8} | {action:>8} | {reward:>6} | {next_state:>10}")
    state = next_state

print(f"\nTotal Reward (Random Policy): ${total_reward}")
```

```python
# greedy policy; maximize immediate reward
state = "medium"
rewards_greedy = 0
for step in range(10):
    action = "release"
    reward = rewards[state][action]
    next_state = transitions[state][action]
    rewards_greedy += reward
    state = next_state

print(f"Total Reward (Greedy): ${rewards_greedy}")

# release only on full

state = "medium"

rewards_strategic = 0
for step in range(10):
    action = "hold" if state != "full" else "release"
    reward = rewards[state][action]
    next_state = transitions[state][action]
    rewards_strategic += reward
    state = next_state

print(f"Total Reward (Strategic): ${rewards_strategic}")
print(f"\nStrategic beats Greedy by: ${rewards_strategic - rewards_greedy}")
# and greedy still beats random
```
