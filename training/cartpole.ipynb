```python
# sacred loop/pattern, skeleton for all rl

obs, info = env.reset(seed=42)
#reset to time 0 and make the first observation


total_reward = 0 #its just return but we cant name it as so cuz return is in-built python fxn
step_count = 0
rewards_per_step = [ ]

terminated = False
truncated = False

while not (terminated or truncated):
    # initially agent decides randomly
    action = env.action_space.sample()
    # here we have two items in the action space
    # go left or go right
    # .sample is simply choosing them randomly


    obs, reward, terminated, truncated, info = env.step(action)
    # this prompt spits out 5 tuple payload

    total_reward+=reward
    step_count+=1
    rewards_per_step.append(reward)

print(f"\nEpisode finished in {step_count} steps")
print(f"Total Reward: {total_reward}")
print(f"Final Observation: {obs}")
print(f"Terminated: {terminated}, Truncated: {truncated}")

env.close()
# close because with only ends the logic here

#Final Observation: [ 0.22001927  1.3644335  -0.24224201 -2.1857803 ]
# array of  [Cart Position, Cart Velocity, Pole Angle, Pole Angular Velocity]

# i see that my episodes were terminated meaning the physics was breached, in this case the pole tilted too much later causing it to free fall with that high angular velo

```

```python
# Run 100 episodes
env = gym.make("CartPole-v1")
episode_rewards = []

for ep in range(100):
    obs, info = env.reset(seed=ep)
    total_reward = 0
    terminated = False
    truncated = False

    while not (terminated or truncated):
        action = env.action_space.sample() # Random placeholder
        obs, reward, terminated, truncated, info = env.step(action)
        total_reward += reward

    episode_rewards.append(total_reward)

env.close()

mean_reward = np.mean(episode_rewards)
std_reward = np.std(episode_rewards)


print(f"Mean Return: {mean_reward:.2f} steps (Max: {np.max(episode_rewards)})")
print(f"Volatility (Std Dev): {std_reward:.2f}")
```
